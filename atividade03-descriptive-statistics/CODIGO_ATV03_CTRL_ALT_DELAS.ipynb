```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
colunas_numericas = df.select_dtypes(include=['int64', 'float64']).columns
print("Colunas numéricas do dataset:")
print(colunas_numericas)
```

```python
print("Resumo estatístico das colunas numéricas:")
df.describe()
```

```python
coluna = df['faturamento_mensal']
print("Analisando a coluna: faturamento_mensal")
```

```python
media = coluna.mean()
mediana = coluna.median()
moda = coluna.mode()[0]
variancia = coluna.var()
desvio_padrao = coluna.std()
amplitude = coluna.max() - coluna.min()
valor_minimo = coluna.min()
valor_maximo = coluna.max()

print("Média:", media)
print("Mediana:", mediana)
print("Moda:", moda)
print("Variância:", variancia)
print("Desvio padrão:", desvio_padrao)
print("Amplitude:", amplitude)
print("Valor mínimo:", valor_minimo)
print("Valor máximo:", valor_maximo)
```

```python
print("INTERPRETAÇÃO INICIAL:")
print(f"A média do faturamento mensal é {media:.2f}.")
print(f"A mediana do faturamento mensal é {mediana:.2f}.")
print(f"A moda do faturamento mensal é {moda:.2f}.")
print(f"O desvio padrão é {desvio_padrao:.2f}, indicando o nível de dispersão dos dados.")
print(f"A amplitude é {amplitude:.2f}, mostrando a diferença entre o maior e o menor faturamento.")
```

```python
plt.figure(figsize=(10,6))
plt.hist(coluna, bins=20, edgecolor='black')
plt.title('Distribuição do Faturamento Mensal')
plt.xlabel('Faturamento Mensal')
plt.ylabel('Frequência')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.boxplot(coluna, vert=False)
plt.title('Boxplot do Faturamento Mensal')
plt.xlabel('Faturamento Mensal')
plt.show()
```

```python
coluna_nota = df['nota_media']

media_nota = coluna_nota.mean()
mediana_nota = coluna_nota.median()
moda_nota = coluna_nota.mode()[0]
variancia_nota = coluna_nota.var()
desvio_padrao_nota = coluna_nota.std()
amplitude_nota = coluna_nota.max() - coluna_nota.min()

print("Estatísticas da coluna nota_media")
print("Média:", media_nota)
print("Mediana:", mediana_nota)
print("Moda:", moda_nota)
print("Variância:", variancia_nota)
print("Desvio padrão:", desvio_padrao_nota)
print("Amplitude:", amplitude_nota)
```

```python
plt.figure(figsize=(10,6))
plt.hist(coluna_nota, bins=15, edgecolor='black')
plt.title('Distribuição da Nota Média')
plt.xlabel('Nota Média')
plt.ylabel('Frequência')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['faturamento_mensal'], df['lucro_estimado'], alpha=0.5)
plt.title('Relação entre Faturamento Mensal e Lucro Estimado')
plt.xlabel('Faturamento Mensal')
plt.ylabel('Lucro Estimado')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
correlacao = df[['faturamento_mensal', 'lucro_estimado', 'numero_pedidos_dia', 'ticket_medio', 'nota_media']].corr()
print("Matriz de correlação:")
print(correlacao)
```

```python
faturamento_por_tipo = df.groupby('tipo')['faturamento_mensal'].mean().sort_values(ascending=False)
print("Média de faturamento por tipo de restaurante:")
print(faturamento_por_tipo)
```

```python
plt.figure(figsize=(12,6))
faturamento_por_tipo.plot(kind='bar')
plt.title('Média de Faturamento por Tipo de Restaurante')
plt.xlabel('Tipo de Restaurante')
plt.ylabel('Faturamento Médio')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
